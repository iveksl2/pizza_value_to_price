```python
import requests 
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
import yaml

# Do not need an Oauth token to access venue data. 
# Get Foursquare Client ID & Client Secret here: https://developer.foursquare.com/ 
with open('config.yml', 'r') as ymlfile:
	config = yaml.load(ymlfile)

def get_venues(foursquare_base_json):
	"""base foursquare explore endpoint json -> venue items hash"""
	return foursquare_base_json['response']['groups'][0]['items']

def get_venue_data(venue_json):
	"""Extracts the name, price & rating from a venue
	   Note: Will not handle missing data
	"""
	name   = venue_json['venue']['name']
	rating = venue_json['venue']['rating']
	price  = venue_json['venue']['price']['tier']
	return {'name': name, 
	        'rating': rating,
	        'price': price}

# The exact lattitude and longitude of the meetup was not used to protect the identity of organizer. For reproducibility these coordinates can be used, obtained via Google (Chicago, IL). Radius argument can also be passed.  
endpoint = 'https://api.foursquare.com/v2/venues/explore?'
payload = {'ll': '41.8781,-87.6298', 
		   'query': 'pizza', 
		   'limit': 25, 
		   'client_id': config['client_id'],
		   'client_secret': config['client_secret'],
		   'v':'20170205'} # update when run. Can make dynamic
```

```python
foursquare_response = requests.get(endpoint, params = payload)
venues = get_venues(foursquare_response.json())
venue_dat = map(get_venue_data, venues)
venues_df = pd.DataFrame(list(venue_dat))
venues_df.set_index(keys = ['name'], drop = True, inplace = True)
venues_df.drop(['Potbelly Sandwich Shop', "Luke's Italian Beef"], inplace = True) # manual
```

```python
venues_df.head(10)
```

### *Higher rated pizza tends to be slightly more expensive* 

```python
sns.jointplot(x="price", y="rating", data=venues_df);
```

```python
# We want to maximize how far our dollar goes
venues_df = venues_df.rename(columns = {'rating': 'value'})
venues_df['value_to_price'] = venues_df['value'] / venues_df['price'] 
venues_df = venues_df.sort_values(by = ['value_to_price'])
```

```python
venues_df['value_to_price'].plot(kind='barh', title="Chicago Pizza's Value to Price Ratio")
```

### Appendix: Possible methodological issues, aka excuses and hedges
- Foursquare `explore` endoint is required to access the rating & price data. However this returns a subset of pre-filtered venues (venues that contain the data). A biased sample by definition. 

- Foursquare Customers are not representative of the Population. Skews heavily towards a younger, tech savy demographic.

- Price likely already partially factored into a `user rating`. Thus dividing by raw `price` may result in an excessive penalty. 

- Price factor is too coarse (1 through 4); i.e. bin width too wide. Simple measure such as average menu price would be better. This is only accessible via foursquare if one is an Owner or a Merchant. 

- People have different utility on quality vs price. (Can expand evaluation criteria and frame as an optimization problem)
